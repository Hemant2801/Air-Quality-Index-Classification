```python
import numpy as np
import pandas as pd
import joblib
```

```python
support = joblib.load('support_vector.sav')
forest = joblib.load('random_forest.sav')
imputer = joblib.load('imputer.sav')
encoder = joblib.load('encoder.sav')
```

```python
df = pd.read_csv('clean.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
x = df.drop('AQI_Bucket', axis = 1)
y = df['AQI_Bucket']
```

```python
x = imputer.transform(x)
x = pd.DataFrame(x, columns = df.columns[0:12])
x.head()
```

```python
x.describe()
```

```python
x.isnull().sum()
```

```python
y = encoder.transform(y)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = .2, stratify = y, random_state = 2)
```

```python
prediction = support.predict(x_train)
print('Support vectore prediction :', prediction)
prediction = forest.predict(x_train)
print('Random forest prediction :', prediction)
print('Actual value :', y_train)
```

```python
prediction = support.predict(x_test)
print('Support vectore prediction :', prediction)
prediction = forest.predict(x_test)
print('Random forest prediction :', prediction)
print('Actual value :', y_test)
```
